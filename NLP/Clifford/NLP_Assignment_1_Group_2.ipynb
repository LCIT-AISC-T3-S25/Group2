```python
# Function to clean text
import unicodedata

def clean_text(text):
    if not isinstance(text, str):
        return ""
    
    # Fix encoding issues and normalize Unicode
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')
    
    # Convert to lowercase
    text = text.lower()
    
    # Expand contractions
    text = contractions.fix(text)
    
    # Replace slang/abbreviations
    text = ' '.join([slang_dict.get(word, word) for word in text.split()])
    
    # Remove URLs while preserving space
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)
    
    # Remove user mentions (@) but keep hashtag text
    text = re.sub(r'@\w+', ' ', text)
    
    # Convert emojis to text WITH SPACES
    text = emoji.demojize(text, delimiters=(" ", " "))
    
    # Handle special punctuation and smart quotes
    text = re.sub(r'[â€œâ€�ð“”‘’\'\"\\]', ' ', text)  # Target problematic chars
    
    # General punctuation handling
    text = re.sub(r'[^\w\s-]', ' ', text)  # Keep hyphens
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
# Visualization functions
def generate_wordcloud(texts, title):
    wordcloud = WordCloud(
        width=800,
        height=400,
        background_color='white',
        stopwords=STOPWORDS,
        max_words=200,
        collocations=False
    ).generate(' '.join(texts))
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, pad=20)
    plt.tight_layout()
    plt.show()

def plot_top_words(texts, n=20, title=None):
    # Tokenize using the imported word_tokenize
    words = [
        word for text in texts 
        for word in word_tokenize(text) 
        if word not in STOPWORDS and len(word) > 2 and word.isalpha()
    ]
    
    # Count and plot
    word_counts = Counter(words)
    top_words = word_counts.most_common(n)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(
        x=[count for word, count in top_words],
        y=[word for word, count in top_words],
        palette='viridis'
    )
    plt.title(title or f'Top {n} Most Frequent Words')
    plt.xlabel('Count')
    plt.ylabel('Word')
    plt.tight_layout()
    plt.show()
```

### iv. Single Layer Unidirectional LSTM using Word2Vec (CBOW/Skipgram)

```python
# Tokenize tweets
df["Tokenized_Tweet"] = df['cleaned_text'].apply(lambda x: word_tokenize(x.lower()))

# Train Word2Vec using CBOW or Skipgram
w2v_model = Word2Vec(
    sentences=df["Tokenized_Tweet"],
    vector_size=embedding_dim,
    window=5,
    min_count=2,
    sg=1,  # sg=1 for Skipgram, sg=0 for CBOW
    workers=4
)

# Convert tweets to average vectors
def tweet_to_vec(tweet, model, target_size=embedding_dim):
    vectors = [model.wv[word] for word in tweet if word in model.wv]
    if not vectors:
        return np.zeros(target_size)
    vec = np.mean(vectors, axis=0)
    return vec

df["Tweet_Vector"] = df["Tokenized_Tweet"].apply(lambda x: tweet_to_vec(x, w2v_model))

# Stack into feature matrix
X_lstm = np.vstack(df["Tweet_Vector"].values)

# Train-test split for Unidirectional LSTM
X_train_lstm, X_test_lstm, y_train_lstm, y_test_lstm = train_test_split(
    X_lstm, y, test_size=0.2, random_state=42, stratify=y
)
```

### Bi-Directional LSTM Model

```python
num_classes = len(np.unique(train_label))
```

```python
# Bi-directional LSTM neural network for sentiment classification
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_len),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(num_classes, activation="softmax")
])
```

```python
# Compile and train the model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",  # For integer-encoded labels
    metrics=["accuracy"]
)

# Train the model using training data and validate on test data
history = model.fit(train_padd_sequence, train_label,
                    validation_data=(test_padd_sequence, test_label),
                    epochs=10)
```

```python
# Visualize training and validation performance over epochs
def plot_graphs(history, attr):
    plt.plot(history.history[attr])
    plt.plot(history.history['val_' + attr])
    plt.xlabel("Epochs")
    plt.ylabel(attr.capitalize())
    plt.title(f"Training vs Validation {attr.capitalize()}")
    plt.legend([attr, 'val_' + attr])
    plt.grid(True)
    plt.show()

# Plot accuracy and loss
plot_graphs(history, "accuracy")
plot_graphs(history, "loss")
```

### Confusion Matrix Built

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predict on test data
y_pred_probs = model.predict(test_padd_sequence)
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# Confusion matrix
cm = confusion_matrix(test_label, y_pred_classes)

# Plot with suppressed scientific notation
fig, ax = plt.subplots(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues', xticks_rotation=45, ax=ax, values_format='d')  # 'd' forces integer display
plt.title("Confusion Matrix - BiLSTM")

# Remove scientific notation from colorbar
ax.figure.axes[-1].yaxis.set_major_formatter(plt.FormatStrFormatter('%d'))

plt.show()
```

### AUC-ROC Curve (One-vs-Rest for Multi-Class)

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc

# Binarize labels and get class-wise ROC curves
n_classes = len(np.unique(test_label))
test_label_bin = label_binarize(test_label, classes=np.arange(n_classes))
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(test_label_bin[:, i], y_pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves
plt.figure(figsize=(8, 6))
colors = ['blue', 'red', 'green', 'orange', 'purple']  # Adjust colors as needed
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-Class ROC Curve (One-vs-Rest)')
plt.legend(loc="lower right")
plt.show()

# Print OvR AUC score
auc_score = roc_auc_score(test_label_bin, y_pred_probs, multi_class='ovr')
print(f"\nOverall AUC Score (OvR): {auc_score:.4f}")
```
