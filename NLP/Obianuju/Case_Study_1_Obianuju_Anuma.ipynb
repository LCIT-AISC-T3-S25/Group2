### Question B-3. Compare the change in word frequencies from stemming and lemmatization. 
#### Which are the top 10 words after stemming/lemmatization?

```python
import sys
import pandas as pd
import re
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from collections import Counter
from nltk.corpus import stopwords
```

```python
# Import the dataset
# sys.path.append('/Users/obianujuanuma/Library/CloudStorage/OneDrive-Personal/Loyalist/Semester 3/Natural Language Processing AISC2009')

corpus = pd.read_csv('Data.csv')
```

```python
corpus.head()
```

```python
len(corpus)
```

```python
# Create a Porter Stemmer Instance
porter_stemmer = PorterStemmer()

# Get all words from the data
words = corpus.iloc[:,0].astype(str).str.cat(sep= ' ') #Concatenate all rows into one string
pattern = r"\b[a-zA-Z][a-zA-Z'-]*[a-zA-Z]\b" #match words including hyphens and apostrophe
# print(words)

print()
# Clean the words
tokens = re.findall(pattern, words.lower())
print(f' Tokenized words: {tokens[:10]}\n')

# Apply stemming to each token
stemmed_words = [porter_stemmer.stem(token) for token in tokens]
print(f'Stemmed words: {stemmed_words[:10]}')
```

```python
len(tokens)
```

```python
len(stemmed_words)
```

```python
# Download WordNet for Lemmatizer
nltk.download('wordnet')
```

```python
# Create lemmatizer instance
lemmatizer = WordNetLemmatizer()
# Apply lemmatization, handling hyphenation
def lemmatize_word(word):
    if '-' in word:
        return '-'.join([lemmatizer.lemmatize(part) for part in word.split('-')])
    else:
        return lemmatizer.lemmatize(word)
    
lemmatized_tokens = [lemmatize_word(token) for token in tokens]
print(f'Lemmatized tokens {lemmatized_tokens[:10]}')
```

```python
len(lemmatized_tokens)
```

```python
# Top 10 words after stemming
top_ten_stems = Counter(stemmed_words).most_common(10)
top_ten_stems
```

```python
# Top 10 words after lemmatization
top_ten_lemmas = Counter(lemmatized_tokens).most_common(10)
top_ten_lemmas
```

```python
# Stemmed words vs Lemmas
top_ten_stems == top_ten_lemmas
```

Since both stemming and lemmatization give the same top 10 numbers, the analysis does not benefit from the complexities of lemmatization. So, it in this case, it is better to go with stemming.
