```python
import pandas as pd
import re
```

```python
corpus = pd.read_csv("C:\\Users\\500236394\\Downloads\\Data.csv")
```

### Question 9

```python
#Iterating over 3 rows of data and replace non alphanumeric with \n
for index, row in corpus.head(3).iterrows(): #only working on 3 rows
    text = row['Data']
    modified_text = re.sub(r'[^a-zA-Z0-9]', '\n', text) # substitute all non alpha numeric with \n
    print(modified_text)
```

### Question 10

```python
url_pattern = r'https?://[^\s]+' #url rejex  pattern
no_of_urls=0    #counter for urls
for index, row in corpus.iterrows():
    text = row['Data']
    no_of_urls += len(re.findall(url_pattern, text)) # adds len of results of search
print(no_of_urls)
```
