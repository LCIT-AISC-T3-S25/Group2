```python
# STEP 1: Load text file without headers (just plain text)
df = pd.read_csv('Downloads/corpus.csv', header=None, names=['text'])
```

```python
# STEP 2: Define your cleaning function
def clean_text(text):
    text = re.sub(r'[^\x00-\x7F]+', ' ', str(text))  # remove non-ASCII
    text = re.sub(r'\s+', ' ', text)                 # normalize whitespace
    return text.strip()
```

```python
# STEP 3: Apply cleaning
df['clean_text'] = df['text'].apply(clean_text)
```

```python
# STEP 4: Drop duplicates based on cleaned text
df_cleaned = df.drop_duplicates(subset='clean_text').reset_index(drop=True)
```

```python
# STEP 5: save result
print(df_cleaned.head())
```

```python
def is_valid_word(word):
    # Check if word starts with an alphabet and is not a URL
    return re.match(r'^[a-zA-Z]', word) and not re.match(r'^(http|https|www)\S*', word)
```

```python
def has_valid_word(text):
    words = text.split()
    return any(is_valid_word(word) for word in words)
```

```python
# Apply to each row to check if it contains at least one valid word
df_cleaned['has_valid_word'] = df_cleaned['clean_text'].apply(has_valid_word)
```

```python
# Count how many rows have at least one valid word
count_valid_records = df_cleaned['has_valid_word'].sum()
```

```python
import pandas as pd
import re

# Regular expression pattern to match decimal numbers
decimal_pattern = r'\d+\.\d+'

# Regular expression pattern to match US-style phone numbers
phone_pattern = r"""
    (?:(?:\+?1[\s.-]?)?         # Optional country code: +1, 1, etc.
    (?:\(?\d{3}\)?[\s.-]?)      # Area code: (123), 123, etc.
    \d{3}[\s.-]?\d{4})          # Final 7 digits of the phone number
"""

# Compile the phone pattern with re.VERBOSE to allow comments and formatting
compiled_phone_pattern = re.compile(phone_pattern, re.VERBOSE)


# Function to extract all decimal numbers from text
def extract_decimals(text):
    return re.findall(decimal_pattern, text)

# Function to extract all phone numbers from text
def extract_phones(text):
    return compiled_phone_pattern.findall(text)

# Function to remove decimal numbers that appear inside phone numbers
def filter_decimals(text, decimals, phones):
    cleaned = []
    for dec in decimals:
        # Keep only those decimals that are NOT part of any phone number
        if not any(dec in phone for phone in phones):
            cleaned.append(dec)
    return cleaned

# Extract phone numbers into a new column
corpus['phone_numbers'] = corpus['Data'].apply(extract_phones)

# Extract all raw decimal numbers into a new column
corpus['raw_decimals'] = corpus['Data'].apply(extract_decimals)

# Remove decimals that are part of phone numbers and store cleaned list
corpus['valid_decimals'] = corpus.apply(
    lambda row: filter_decimals(row['Data'], row['raw_decimals'], row['phone_numbers']),
    axis=1
)

# Count how many valid decimal numbers exist in each row
corpus['decimal_count'] = corpus['valid_decimals'].apply(len)

# Filter the rows where at least one valid decimal was found
corpus_filtered = corpus[corpus['decimal_count'] > 0][['Data', 'valid_decimals']]

# Print the filtered rows in a readable format
for idx, row in corpus_filtered.iterrows():
    print(f"Row index {idx} contains decimals: {row['valid_decimals']}")
    print(f"Line: {row['Data']}\n")
    print("-------------------------------------------------------")
```

## Question B.2.

```python
# Import the necessary libraries
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

import unicodedata
import string
import re
```

```python
# Import the corpus
corpus = pd.read_csv('./Data.csv')
```

```python
# Check the first 5 documents
corpus.head()
```

```python
# Function to clean corpus
def preprocess(word):
    # Remove URLs
    word = re.sub(r'http\S+|www\S+|https\S+', '', word, flags=re.MULTILINE)
    # Remove '@' and hash '#' tags
    word = re.sub(r'@\w+', '', word)
    word = re.sub(r'#', '', word)
    word = re.sub(r'\brt\b', '', word, flags=re.IGNORECASE)
    # normalize unicode to ascii
    word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8')
    # Remove standalone digits
    word = re.sub(r'\b\d+\b', '', word)
    # Casefold
    word = word.casefold()
    # Normalize whitespace
    word = ' '.join(word.split())
    # Remove some special characters while maintaining necessary ones
    word = re.sub(r"[^\w\s'-]", '', word)

    # handle slashes between words
    if '/' in word:
        # first split to maintain known combinations
        if word.lower() in {'a/b', 'w/d', 'w/o', 'b/c'}:
            return word.lower()
        # second split
        parts = word.split('/')
        if len(parts) == 2 and len(parts[0]) < 8 and len(parts[1]) < 8:
            return '_'.join(parts)  # mountain_river
        # third split for other cases
        return ' '.join(parts)

    return word
```

```python
# Apply cleaning to corpus
corpus['clean_data'] = corpus['Data'].apply(preprocess)
```

```python
# Tokenize
corpus['tokens'] = corpus['clean_data'].apply(word_tokenize)
```

```python
# Total count of words before lemmatization
all_tokens = [token for lst in corpus['tokens'] for token in lst]
num_unique_before = len(set(all_tokens))

print(f'Number of unique words before lemmatization: {num_unique_before}')
```

```python
# Lemmatizer initialization
lemmatizer = WordNetLemmatizer()
```

```python
# Lemmatization function
def lemmatize_text(word):
    tokens = word_tokenize(word)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)
```

```python
# Count unique words after lemmatization
corpus['lemmas'] = corpus['clean_data'].apply(lemmatize_text)
all_lemmas = ' '.join(corpus['lemmas']).split()
num_unique_after = len(set(all_lemmas))

print(f'Number of unique words before lemmatization: {num_unique_after}')
```

```python
print(f"Reduction: {num_unique_before - num_unique_after} words")
```

## Question B.3.

##### Compare the change in word frequencies from stemming and lemmatization. 
##### Which are the top 10 words after stemming/lemmatization?

```python
import pandas as pd
import re
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from collections import Counter
from nltk.corpus import stopwords
```

```python
# Create preprocess function
def preprocess_text(word):
    """Lowercase, remove punctuation and tokenize"""
    word = word.lower()
    # Remove characters that are not letters, apostrophes, white spaces or hyphens
    word = re.sub(r"[^a-zA-Z'\s-]", '', word)
    tokens = word_tokenize(word)
    return tokens
```

```python
# Process the data
def process_data(file_path, text_column, remove_duplicates=True):
    """Load CSV file, handle duplicates, and return tokens"""
    corpus = pd.read_csv(file_path)
    if remove_duplicates:
        corpus = corpus.drop_duplicates()
    word = ' '.join(corpus[text_column].astype(str))
    return preprocess_text(word)
```

```python
# Determine frequencies
def analyze_frequencies(tokens):
    """Compare stemmed vs lemmatized frequencies"""
    # Create stemmer and lemmer instances
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    # Original frequencies
    original_freq = Counter(tokens)

    # Stemmed frequencies
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    stemmed_freq = Counter(stemmed_tokens)

    # Lemmatized frequencies
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    lemmatized_freq = Counter(lemmatized_tokens)

    return original_freq, stemmed_freq, lemmatized_freq
```

```python
# Print comparisons

def print_comparison(original, stemmed, lemmatized):
    """Print top comparisons"""
    print("Unique Terms Count:")
    print(f"Original: {len(original)}")
    print(f"Stemmed: {len(stemmed)} (Difference: {len(original) - len(stemmed)})")
    print(f"Lemmatized: {len(lemmatized)} (Difference: {len(original) - len(lemmatized)})")

    print(f"\nTop 10 Original terms: {original.most_common(10)}")
    print(f"Top 10 Stemmed terms: {stemmed.most_common(10)}")
    print(f"Top 10 Lemmatized terms: {lemmatized.most_common(10)}")
   
```

```python
# Importing required libraries
import pandas as pd
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from collections import Counter
import re
```

```python
# Load CSV file provided for assignment
from google.colab import files
uploaded = files.upload()
```

```python
# Taking the uploaded file
file_name = list(uploaded.keys())[0]
corpus = pd.read_csv(file_name)
```

```python
# Using Preprocessing tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
# Clean and tokenize Function
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    return tokens
```

```python
# Removing stop words Function
def remove_stop_words(words):
    return [word for word in words if word not in stop_words]
```

```python

# Stemming function
def stem_words(words):
    return [stemmer.stem(word) for word in words]
```

```python
#Lemmatization Function
def lemmatize_words(words):
    return [lemmatizer.lemmatize(word) for word in words]

```

```python
# Flatten helper Function
def flatten(list_of_lists):
    return [item for sublist in list_of_lists for item in sublist]
```

```python
# Preprocessing pipeline
tokens = corpus['Data'].dropna().apply(preprocess)
```

```python

# Normalization technique only (before stop word removal)
stemmed_only = tokens.apply(stem_words)
lemmatized_only = tokens.apply(lemmatize_words)
```

```python
#Normalization technique after stop word removal
filtered = tokens.apply(remove_stop_words)
stemmed_filtered = filtered.apply(stem_words)
lemmatized_filtered = filtered.apply(lemmatize_words)
```

```python
# Flattennig all
flat_stemmed_only = flatten(stemmed_only)
flat_lemmatized_only = flatten(lemmatized_only)
flat_stemmed_filtered = flatten(stemmed_filtered)
flat_lemmatized_filtered = flatten(lemmatized_filtered)
```

```python
# Counting word frequencies
freqs = {
    'Stemming only': Counter(flat_stemmed_only).most_common(20),
    'Lemmatization only': Counter(flat_lemmatized_only).most_common(20),
    'Stemming after stop word removal': Counter(flat_stemmed_filtered).most_common(20),
    'Lemmatization after stop word removal': Counter(flat_lemmatized_filtered).most_common(20)
}
```
