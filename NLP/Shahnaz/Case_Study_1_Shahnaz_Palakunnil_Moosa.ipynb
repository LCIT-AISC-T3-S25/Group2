```python
import pandas as pd
import re
from nltk.stem import PorterStemmer

# Load the CSV file
# Import the corpus
corpus = pd.read_csv('C:\\PG AI and DS\\Semester 3\\NLP AISC2009\\Data.csv')

# Check the first 5 documents
corpus.head()

# Combine all text into a single string
text = ' '.join(corpus['Data'].astype(str).tolist())

# Tokenize using regex
tokens = re.findall(r'\b[a-zA-Z]+\b', text.lower())

# Unique tokens before stemming
unique_tokens_before = set(tokens)
count_before = len(unique_tokens_before)
print("Unique tokens before stemming:", count_before)

# Apply Porter Stemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokens]

# Unique tokens after stemming
unique_tokens_after = set(stemmed_tokens)
count_after = len(unique_tokens_after)
print("Unique tokens after stemming:", count_after)

# Optional: show some examples
for original, stemmed in zip(tokens[:10], stemmed_tokens[:10]):
    print(f"{original} → {stemmed}")
```

```python
# Import necessary libraries

import pandas as pd

from nltk.tokenize import word_tokenize

from nltk.stem import PorterStemmer

from nltk.corpus import stopwords

import string

import nltk
 
# Download required NLTK data

nltk.download('punkt')

nltk.download('stopwords')
 
# Import the corpus
corpus = pd.read_csv('C:\\PG AI and DS\\Semester 3\\NLP AISC2009\\Data.csv')

# Check the first 5 documents
corpus.head()
 
# Combine all text data into a single string (adjust the column name as needed)

text_data = " ".join(corpus.astype(str).apply(" ".join, axis=1))
 
# Tokenize the text

tokens = word_tokenize(text_data)
 
# Remove punctuation and make lowercase

tokens = [word.lower() for word in tokens if word.isalpha()]
 
# Count unique tokens before stemming

unique_tokens_before = set(tokens)

print(f"Number of unique tokens before stemming: {len(unique_tokens_before)}")
 
# Apply Porter Stemmer

stemmer = PorterStemmer()

stemmed_tokens = [stemmer.stem(word) for word in tokens]
 
# Count unique tokens after stemming

unique_tokens_after = set(stemmed_tokens)

print(f"Number of unique tokens after stemming: {len(unique_tokens_after)}")

```
