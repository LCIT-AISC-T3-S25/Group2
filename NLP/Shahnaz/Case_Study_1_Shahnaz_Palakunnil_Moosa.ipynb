```python
import pandas as pd
import re
from nltk.stem import PorterStemmer

# Load the CSV file
df = pd.read_csv('C:\\PG AI and DS\\Semester 3\\NLP AISC2009\\Data.csv')

# Combine all text into a single string
text = ' '.join(df['Data'].astype(str).tolist())

# Tokenize using regex
tokens = re.findall(r'\b[a-zA-Z]+\b', text.lower())

# Unique tokens before stemming
unique_tokens_before = set(tokens)
count_before = len(unique_tokens_before)
print("Unique tokens before stemming:", count_before)

# Apply Porter Stemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokens]

# Unique tokens after stemming
unique_tokens_after = set(stemmed_tokens)
count_after = len(unique_tokens_after)
print("Unique tokens after stemming:", count_after)

# Optional: show some examples
for original, stemmed in zip(tokens[:10], stemmed_tokens[:10]):
    print(f"{original} → {stemmed}")
```

```python
import pandas as pd
import re
from nltk.stem import PorterStemmer

# Load dataset
df = pd.read_csv('C:\\PG AI and DS\\Semester 3\\NLP AISC2009\\Data.csv')

# Prepare and clean text
text = ' '.join(df['Data'].dropna().astype(str)).lower()

# Tokenization using regex (alphabetic tokens only)
tokens = re.findall(r'\b[a-z]+\b', text)

# Unique tokens before stemming
unique_before = set(tokens)

# Initialize stemmer and apply stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in tokens]

# Unique tokens after stemming
unique_after = set(stemmed)

# Output results
print(f"Unique tokens before stemming: {len(unique_before)}")
print(f"Unique tokens after stemming: {len(unique_after)}")

# Show sample transformations
print("\nSample token transformations:")
for original, stemmed_word in zip(tokens[:10], stemmed[:10]):
    print(f"{original} → {stemmed_word}")
```
