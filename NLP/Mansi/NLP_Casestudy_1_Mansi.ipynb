```python
import pandas as pd

file_path = "/content/drive/MyDrive/Data/Data.csv"
df = pd.read_csv(file_path)
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nSample Data:")
print(df.head())
```

```python
data_pattern = r'\b(?:\d{1,2}[\/\-\.]){2}\d{2,4}\b|\b\d{4}[\/\-\.](?:\d{1,2}[\/\-\.]){2}\b'
```

```python
import re
df['Has_Number'] = df['Data'].apply(lambda x: bool(re.search(data_pattern, x)))
```

```python
data_count = df['Has_Number'].sum()
```

```python
print(f"Number of records containing any number (no alphabets): {data_count}")
```

Number of records have a word starting with the letter “w”

```python
w_word_pattern = r'\bw\w*'
```

```python
df['Has_W_Word'] = df['Data'].apply(lambda x: bool(re.search(w_word_pattern, x, re.IGNORECASE)))
```

```python
w_word_count = df['Has_W_Word'].sum()
```

```python
print(f"Number of records containing a word starting with 'w': {w_word_count}")
```
